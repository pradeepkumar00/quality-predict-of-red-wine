## Creating pipeline

```python
train_set_tr = train_set.drop("quality", axis=1)
train_set_labels = train_set['quality'].copy()
```

## Choose desired model

1st model

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(train_set_tr, train_set_labels)
```

```python
some_data = train_set_tr.iloc[:5]
some_lables = train_set_labels.iloc[:5]
some_lables
```

```python
predict_value = model.predict(some_data)
```

```python
predict_value
```

```python
list(some_lables)
```

## Evaluating model

```python
import numpy as np
from sklearn.metrics import mean_squared_error
data_predictions = model.predict(train_set_tr)
mse = mean_squared_error(train_set_labels, data_predictions)
rmse = np.sqrt(mse)
```

```python
rmse
```

2nd model

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
model.fit(train_set_tr, train_set_labels)
```

```python
predict_val_dtree = model.predict(some_data)
```

```python
predict_val_dtree
```

```python
list(some_lables)
```

# Cross Validation for Decision tree

```python
from sklearn.model_selection import cross_val_score
val_sc = cross_val_score(model, train_set_tr, train_set_labels, scoring ="neg_mean_squared_error", cv=10)
rmse_sc = np.sqrt(-val_sc)
```

```python
rmse_sc
```

```python
def print_scores(scores):
    print("Scores: ", scores)
    print("Mean: ", scores.mean())
    print("Standard deviation: ", scores.std())
```

```python
print_scores(rmse_sc)
```

#Linear_regrassion is better than Decision tree

3rd Model

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(train_set_tr, train_set_labels)
```

```python
predict_val_rfr = model.predict(some_data)
```

```python
predict_val_rfr
```

```python
list(some_lables)
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error
data_predictions_rfr = model.predict(train_set_tr)
mse = mean_squared_error(train_set_labels, data_predictions_rfr)
rmse_rfr = np.sqrt(mse)
```
